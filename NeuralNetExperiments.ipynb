```python
# PARAMS FOR ALL NETWORKS
# sine
sine_learning_rate = 0.01
sine_no_of_epochs = 1000 
# normal
normal_learning_rate = 0.001
normal_no_of_epochs = 1000 
# all
dropout_rate = 0.0 
```

```python
sine_inputs, sine_outputs = file_io.get_dataset("./data/sine")
# print("sine_inputs=", sine_inputs)
# print("sine_outputs=", sine_outputs)
```

```python
deep_sine_qlearn = neural_factory.get_neural_net(no_of_inputs=1, no_of_neurons_layer1=25, no_of_neurons_layer2=2,
                                     learning_rate=sine_learning_rate, network_name="sine_qlearn_deep_model")
sine_inputs, sine_outputs = file_io.get_dataset("./data/sine")

predicted_sine_outputs_layer2, net_sine_losses_layer2, \
    net_sine_losses_layer1, net_sine_losses_layer0, \
    param_sine_conditioning_layer2, param_sine_conditioning_layer1, \
    large_singular_value_sine_layer1, small_singular_value_sine_layer1, \
    large_singular_value_sine_layer2, small_singular_value_sine_layer2 = \
        dataset_gen.data_default_gathering(sine_inputs, sine_outputs, deep_sine_qlearn, \
                  no_of_epochs=sine_no_of_epochs) 
```

```python
sine_accumulate_outputs = [row[0] for row in sine_outputs.astype(float)]
sine_distribute_outputs = [row[1] for row in sine_outputs.astype(float)]
sine_accumulate_datset = np.column_stack((np.array(sine_inputs).astype(float), \
                                          np.array(sine_accumulate_outputs).astype(float)))
sine_distribute_datset = np.column_stack((np.array(sine_inputs).astype(float), \
                                          np.array(sine_distribute_outputs).astype(float)))
print("sine_accumulate_datset[0]=", sine_accumulate_datset[0])
print("sine_distribute_datset[0]=", sine_distribute_datset[0])
```

```python
predicted_sine_accumulate_layer2 = [[row[0]] for row in np.array(predicted_sine_outputs_layer2).astype(float)]
evaluated_sine_accumulate_dataset = np.column_stack((np.array(sine_inputs).astype(float), \
                                                     predicted_sine_accumulate_layer2))
print("evaluated_sine_accumulate_dataset[0]=", evaluated_sine_accumulate_dataset[0])
plot_deep.plot_multi_line(sine_accumulate_datset, 'true output', evaluated_sine_accumulate_dataset, 'predicted output', \
                    'training (Accumulate Sine QLearning )', 'input', 'output')
```

```python
predicted_sine_distribute_layer2 = [[row[1]] for row in np.array(predicted_sine_outputs_layer2).astype(float)]
evaluated_sine_distribute_dataset = np.column_stack((np.array(sine_inputs).astype(float), \
                                                     predicted_sine_distribute_layer2))
print("evaluated_sine_distribute_dataset[0]=", evaluated_sine_distribute_dataset[0])
plot_deep.plot_multi_line(sine_distribute_datset, 'true output', evaluated_sine_distribute_dataset, 'predicted output', \
                    'training (Distribute Sine QLearning )', 'input', 'output')
```
