```python
# https://docs.scipy.org/doc/numpy-1.15.0/reference/generated/numpy.random.normal.html
min_value = -6
max_value = 12
dist_range = -min_value+max_value
num_samples = 50000
```

```python
### NORMAL DISTRIBUTION
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

standard_mean = 3
standard_deviation = 2
normal_distribution = norm(loc=standard_mean, scale=standard_deviation)
```

```python
x = np.linspace(min_value, max_value, num_samples)

plt.plot(x, normal_distribution.pdf(x), '-', lw=2)
```

```python
### DRAW FROM NORMAL DISTRIBUTION
normal_draw = np.random.normal(loc=standard_mean, scale=standard_deviation, size=num_samples)

num_buckets = 30
plt.hist(normal_draw, num_buckets, density=True)
```

```python
output_file = "normal_distribution"
np.save(output_file, normal_draw)
np.load(output_file+".npy")
```

```python
plt.plot(x, normal_draw, '-', lw=2)
```

```python
### SINE DISTRIBUTION
import matplotlib.pyplot as plt
import numpy as np


# https://en.wikipedia.org/wiki/Sine_wave
wave_length = 100000 # 1000
oscillation_frequency = 5
max_time = num_samples*10 
time_ticks = np.arange(max_time)
def price_ticks(time_ticks):
    angular_frequency = 2 * np.pi * oscillation_frequency
    time_in_wave = time_ticks/wave_length
    sine_wave = np.sin(angular_frequency*time_in_wave)
    transformed_sine = sine_wave*standard_deviation + standard_mean
    return transformed_sine
```

```python
plt.xlabel('time(t)')
plt.ylabel('price($)')
sine_prices = price_ticks(time_ticks)
plt.hist(sine_prices, num_buckets, density=True)
```

```python
plt.plot(time_ticks, sine_prices)
```

```python
sine_draw = np.copy(sine_prices)
np.random.shuffle(sine_draw)
plt.plot(time_ticks, sine_draw)
```

```python
output_file = "sine_distribution"
np.save(output_file, sine_draw)
np.load(output_file+".npy")
```
